## Check Missing Values

```python
# Check whether missing values(na) exist
data.isna().sum()
```

isna() 결과, pandas.DataFrame.describe()에서 확인할 수 있는 column별 min, max값을 확인해, 위에서 제시한 범주형 변수에는 결측치가 없는 것을 확인할 수 있습니다.

중복되는 행은 분포 왜곡을 유발하고 추후 모델 학습에 사용 시 성능에 영향을 미칠 수 있습니다. 중복되는 행이 있는지 확인하고, 있다면 삭제합니다.

```python
# Remove duplicated rows
data[data.duplicated()]
```

모든 column에 대해 일치하는 행은 없음을 확인할 수 있습니다. \
그러나, 한 명의 직원에 대해 모순되는 여러 데이터가 존재하는 등의 문제가 존재할 수 있습니다.

예를 들어, Relation Extraction에서 서로 다른 두 행이 문장과 같은 entity pair를 가지고 있으나, entity에 대한 label이 다른 경우, 학습 시 모델이 혼동할 수 있습니다.

중복되는 직원 데이터가 있는지 확인합니다.

```python
# Check the number of unique values of each columns
[(column, len(set(data[column].unique()))) for column in data.columns.values.tolist()]
```

직원을 구분하기 위해 사용하는 EmployeeNumber는 총 1470개의 unique value를 가지고, 본 데이터는 총 1,470개의 행으로 이루어져 있으니, 위와 같이 같은 직원인데 다른 값을 가지는 데이터는 없음을 확인할 수 있습니다.

또한 unique value 확인을 통해, 값이 오로지 한 개인 column도 발견할 수 있었습니다! \
Unique value가 row 수와 같은 `EmployeeNumber`, unique value가 오직 한 개인 `EmployeeCount`, `Over18`, `StandardHours`는 불필요하므로 제거합니다.

```python
# Remove unnecessary features
unnecessary_columns = ["EmployeeNumber", "EmployeeCount", "Over18", "StandardHours"]
data.drop(columns=unnecessary_columns, inplace=True)
numerical_discrete_features = list(set(numerical_discrete_features) - set(unnecessary_columns))
```

pandas.DataFrame.describe()를 통해 간단히 확인했을 때, DailyRate와 같은 경우 표준편차가 매우 컸습니다. 이상치는 없는지 확인합니다.

우선 히스토그램을 그려 numerical discrete data 중 이상치가 없는지 직관적으로 확인합니다.


```python
data[numerical_discrete_features].hist(figsize=(15, 15))
plt.title("Boxplot - Numerical Discrete Features")
plt.show()
```

히스토그램으로 한눈에 보기 어려운 부분을 다시 보기 위해, numerical discrete data에 대해 describe를 활용합니다.

```python
data[numerical_discrete_features].describe()
```

`YearsWithCurrManager`와 `YearsAtCompany`는 높은 상관관계를 보이지만, 생각해보면 자연스러운 결과입니다! \
이 회사에 있는 시간과, 현재 매니저와 함께 있는 시간은 비례할 가능성이 매우 높습니다.

`TotalWorkingYears`와 `MonthlyIncome`간의 상관관계도, 오래 근무할수록 급여가 상승된다는 당연한 관계를 표현합니다.

다른 변수들과 `Attrition`간의 관계를 살펴봅니다.

```python
attrition_as_num = data["Attrition"].replace(["Yes", "No"], [1, 0], inplace=False)
plt.figure(figsize=(15, 10))
age_attrition_barplot = sns.barplot(x=data["Age"],y=attrition_as_num)
age_attrition_barplot.set_title(f"Age - Attrition")
plt.show()
```

`Age`와 `Attrition`의 관계를 살펴봅니다.
18~20대 초반, 50대 후반~에서 퇴직률이 높은 것을 확인할 수 있습니다.

```python
len(data["MonthlyIncome"].unique())
```

`MonthlyIncome`은 unique value가 너무 많으니 다른 방식으로 시각화해봅시다.

```python
plt.figure(figsize=(15, 10))
income_attrition_barplot = sns.violinplot(y=data["MonthlyIncome"], x=attrition_as_num)
income_attrition_barplot.set_title(f"Income - Attrition")
plt.show()
```

급여가 많으면 퇴직 확률이 낮음을 직관적으로 확인할 수 있습니다.
